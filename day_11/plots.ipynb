# PLOTING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kashti = sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived', hue='sex', data=kashti)
```

```python
# Create a scatter plot of the 'age' and 'fare' columns
plt.scatter(kashti['age'], kashti['fare'])

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare')

# Show the plot
plt.show()
```

```python
# scatter plot

# plt.scatter(data=kashti, x='age', y='fare', hue='pclass')
# If you want to color the points based on a third column, such as pclass, you can use the c parameter of the scatter function to specify the column to use for color. For example:
plt.scatter(kashti['age'], kashti['fare'], c=kashti['pclass'])
```

```python


sns.scatterplot(data=kashti, x='age', y='fare', hue='pclass')
```

```python
sns.barplot(data=kashti, x='sex', y='fare', hue='pclass')
```

```python
kashti.columns
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins=10, figsize=(9,7), grid=False)
```

```python
sns.boxplot(data=kashti, x='sex', y='fare', hue='class')
```

```python
sns.boxenplot(data=kashti, x='sex', y='fare', hue='pclass')
```

```python
g = sns.FacetGrid(kashti, hue='survived', col= 'pclass', margin_titles=True, palette={1:'seagreen', 0:'gray'})

g=g.map(plt.scatter, 'fare', 'age', edgecolor='w').add_legend()
```

```python
g = sns.FacetGrid(kashti, hue='survived', col= 'who', margin_titles=True, palette={1:'seagreen', 0:'gray'})

g=g.map(plt.scatter, 'fare', 'age', edgecolor='w').add_legend()
```

```python
sns.stripplot(data=kashti, x='sex', y='fare', hue='pclass')
```

```python
#heatmap

corr = kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
```

```python
sns.violinplot(data=kashti, x='sex', y='fare', hue='pclass')
plt.title("VIOLIN PLOT")
```
