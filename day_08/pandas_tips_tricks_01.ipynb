# 16- Spliting string into multiple columns

```python
import pandas as pd

df = pd.DataFrame({'name':['Arslan khan', 'Luqman Blti', 'Nomi don' ,'Hamza bakrii'], 'location': ['lahore, Pak', 'Pindi, Pak', 'Isl, Pak', 'Quetta, Pak']})

df
```

```python
#spliting a column in to two

# df = df.name.str.split(' ', expand=True)

```

```python
# adding those split in to new column

df[['first_name', 'last_name']] = df.name.str.split(' ', expand=True)
df
```

```python
df[['city', 'country']] = df.location.str.split(',', expand=True)
df
```

```python
# refins data manipulation
df =df[['first_name', 'last_name', 'city', 'country']]
df
```
